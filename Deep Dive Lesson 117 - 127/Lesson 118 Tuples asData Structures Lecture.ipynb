```python
'''
Tuples vs Lists vs Strings 

Tuples                               Lists                                         Strings
containers                           containers                                    containers 

order matters                        order matters                                 order matters       Sets' orders do not matter
Heterogeneous(usually)/Homogeneous   Heterogeneous/Homogeneous(usually)            Homogeneous

indexable                            indexable                                     indexable 
iterable                             iterable                                      iterable 

Immutable                            Mutable                                       Immutable 
     fixed length                         length can change                             fixed length 
     fixed order                          order of elements can change                  fixed order 
          cannot do in-place sorts                 can do in-place sorts                      cannot do in-place sorts
          cannot do in-place reversals             can do in-place reversals                  cannot do in-place reversals




Immutability of Tuples 

Elements cannot be added or removed 
the order of elemnts cannot be changed 
works well for representing data structures: 

the meanings of positions

point: (10, 20)                1st element is the x-coordinate
                               2nd element is the y-coordinate

circle: (0,0,10)               1st element is the x-coordinate of the center 
                               2nd element is the y-coordinate of the center 
                               3rd element is the radius 


city: ('London', 'UK', 8_780_000)   1st element is the name of a city
                                    2nd element is the country
                                    3rd element is the population 
                                    
The position of the data has meaning 

TUPLES as data records 

Think of a tuple as a data record where the position of the data has meaning 

london = ('London', 'UK', 8_780_000)
new_york = ('New York', 'USA', 8_850_000)
beijing = ('Beijing', 'China', 21_000_000) 


Because tuples, strings and integers are immutable, we are guaranteed that the data ana data structure for 
London will never change. 

We can have a list of these tuples: 

cities = [('London', 'UK', 8_780_000),
          ('New York', 'USA', 8_850_000), 
          ('Beijing', 'China', 21_000_000]


Extracting data from Tuples 

Since tuples are sequences just like strings and lists, we can retrieve items by index 

london = ('London', 'UK', 8_780_000)

city = london[0] country = london[1] population = london[2]

cities = [('London', 'UK', 8_780_000),
          ('New York', 'USA', 8_850_000),    The tuple is heterogeneous. However, the list is homogeneous. 
          ('Beijing', 'China', 21_000_000]
          
total_population = 0 
for city in cities: 
    total_population += city[2]

The list of cities is homogeneous (contain cities only)
However, a city (the tuple) is heterogeneous 

Extracting data from Tuples 

We can also use tuple unpacking 
we actually already know ho wto do this == we covered this in the section on function arguments 
new_york = ('New York', 'USA', 8_850_000) -----> packed three values into a tuple 

city, country, population = new_york -----> unpacking the tuple 

process: 
first element of new york assign to city, second to 
country, and third to population 


city, country, population = ('New York', 'USA', 8_500_000)
city, country, population = 'New York', 'USA', 8_500_000   ------> These are the same () can further diffeciate tuples though. 


Dummy Variables 

This is something you are likely to run across when you look at Python code that uses tuple unpacking 
Sometimes, we are only interested in a subset of the data fileds in a tuple, not all of them 

Suppose we are interested in only the city name, and the population: 

city, _, population == ('Beijing', 'China', 21_000_000) 

_is  actualy a legal name -- so there is nothing special about it 

but by onvention, we use the underscore to indicate this is a variable we don't care about 

in fact, we could just have used: 

city, ignored, population = ('Beijing', 'China', 21_000_000)


Dummy Variables 

It's also used in extended unpacking too 
record = ('DJIA', 2018, 1, 19, 25987.35, 26071.71, 25942.83, 26071,72)
symbol, year, month, day, ooen, high, low close = record 

Let's say we are only interested in the symbol, year, month, day, and close fields 

We could do it this way: 

symbol = record[0] 
year = record[1]
month = record[2]
day = record[3]
close = record[7]

symbol, year, close = record[0], record[1], record[7]

symbol, year, month, day, *_, close = record 
symbol, year, month, day, *ignored, close = record 


'''
```
