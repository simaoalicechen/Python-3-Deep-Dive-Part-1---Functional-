```python
'''
Functional Equivalents to Operator 
In the last lecture we wrote code such as: 

l = [2,3,4]
reduce(lambda a,b: a*b, l)
We used a lambda expression to create a funtional version of the * operator 
This is something that happens quite often, so the operator modulewas created 

This moduel is a convenience module. 

you Can always use your own functions and lambda expressions instead. 


The operator module 

Arithmetic Functions 

add(a,b)
mul(a,b)
pow(a,b)
mod(a,b)
floordiv(a,b)
neg(a,n)
....



Comparison and Boolean Operators 
lt(a,b)
le(a,b)
gt(a,b)
ge(a,b)
eq(a,b)
ne(a,b)
is_(a,b)
is_not(a,b)
and_(a,b)
or_(a,b)
not_(a,b)  _ to differentiate from the built in operator 

concat(s1,s2) 
contains(s,val) 
contOf(s, val) 
getitem(s,i) is an index 
setitem(s,i,val) 
delitem(s, i) ---> only when s is iteratable

variants that use slices for the last three


Item Getters 

The itemgetter function returns a callable

getitem(s, i)  takes two parameters, and returns a value: s[i] 

itemgetter(i)  --> partial function, only takes the index returns a callablle which takes one parameter: a sequence object 
f = itemgetter(1) 
s = [1,2,3] 
f(2) = 3 

Item Getters
We can pass more than one index to itemgetter: 

l = [1,2,3,4,5,6] 
s = 'python' 

f = itemgetter(1,3,4) 
f(l)  ---> (2,4,5) 
f(s)  ---> ('y', 'h', 'o') 

Attribute Getters 

The attrgetter function is similar to itemgetter, but is used to retrieve object attributes 

It also returns a callable, that takes the object as an argument 

Supporse my_obj is an object with three properties: 

my_obj.a -->10 
my_obj.b -->20 
my_obj.c -->30

f = attrgetter('a') f(my_obj)   ---> 10 

f = attrgetter('a', 'c')  f(my_obj) -->(10,30) 
Can also call directly: 

attrgetter('a', 'b', 'c')(my_obj) 


Calling another Callable 

Consider the str class that provides the upper() method: 

s = 'python' s.upper() = PYTHON 
f = attrgetter('upper') 

f(s) -->returns the upper method of s 

f(s)() -->PYTHON 
attrgetter('upper')(s)() 

Or, we can use the slightly simpler 'methodcaller' function 
methodcaller('upper')('python') 

Basically, methodcaller retrieves the named attribute and calls it as well. 
It can also handle more arguments, as we'll in the code





'''
```
