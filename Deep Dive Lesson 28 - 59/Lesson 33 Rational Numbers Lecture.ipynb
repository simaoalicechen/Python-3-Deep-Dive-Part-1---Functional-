```python
'''
Rational numbers are fractions of integer numbers 
ex: 1/2, -22/7

Any real number with a finite number of digits after the decimal point is also a rational number 
0.45 --> 45/100   0.123456789 ---> 123456789/10**9
So 8.3/4 is also rational 8.3/4 = 83/10/4=83/40
as is 8.3/1.4 since 83/14

The Fraction Class 
Rational numbers can be represented in Python using the Fraction class in the fractions module 
from fractions import Fraction 
x = Fraction(3,4)
y = Fraction(22,7)
z = Fraction(6,10)

Fractions aare automatically reduced: 
 Fraction(6, 10) --> Fraction(3,5) 
 
Negative sign, if any, is always attached to the numerator: 
Fraction(1,-4) ---> Fraction(-1, 4)

Constructors 
Fraction(numberator = 0, denominator = 1)
Fraction(other_fraction)
Fraction(float)
Fraction(decimal)
Fraction(string) 

Fraction('10')  --> Fraction(10,1)
Fraction('0.125') --> Fraction(1,8)
Fraction('22/7') --> Fraction(22,7) 

Standard arthmetic operators are supported: +, -,*, /
and result in Fraction objects as well 

(2/3)*(1/2) = 2/6 = 1/3 Fraction(2,3)* Fraction(1,2) --> Fraction(1,3)
(2/3)+(1/2) = 7/6 = 1/3 Fraction(2,3)+ Fraction(1,2) --> Fraction(7,3)
Getting the numerator and denominator fo Fraction objects: 
x = Fraction(22,7)
x.numerator -----> 22
x.denominator -----> 7 

floats objects have finite precision (you can only use so many digits to present certain numbers)

=====> any float object can be written as a fraction! (They are actually Rational numbers!) 

Fraction(0.75)  ---> Fraction(3,4)
Fraction(1.375) ---> Fraction(11,8)

import math 
x = Fraction(math.pi) Fraction(.......,.......) a proximation of the irrational number 

Converting a float to a Fraction has an important caveat

1/8 has an exact float representation 
Fraction(0.125) --> Fraction(1,8)
3/10 does not have an exact representation 
Fraction(0.3) ---> Fraction(.....,.....)
format(0.3, 'o.5f') ----> 0.300000
format(0.3, '0.25f') ----> 0.299999999989

Constraining the denominator 

Given a Fraction object, we can fina an approximate equivalent fraction with a constrained denomintaor 

i.e finds the closet rational(which could be precisely equal)
with a denominator that does not exceed max_denominator 

x=Fraction(math.pi)

x.limit_denominator(10) ---> Fraction(22,7) 

x.limit_denominator(100) ---> Fraction(311,99)

x.limit_denominator(500) ---> Fraction(355,113) 


'''
```
